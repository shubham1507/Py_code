Some data types are inhehrited from Python's native types:

| Data type | Tipo nativo |
|---|---|
|`int_`|`int`|
|`float_`|`float`|
|`complex_`|`complex`|
|`bytes_`|`bytes`|
|`str_`|`str`|
|`bool_`|`bool`|
|`datetime64`|`datetime.datetime`|
|`timedelta64`|`datetime.timedelta`|

Some data types have *aliases*:

Example:

| Alias | Nome original |
|---|---|
|`np.int16`|`np.short`|
|`np.int32`|`np.intc`|
|`np.int64`|`np.int_`|
|`np.float16`|`np.half`|
|`np.float32`|`np.single`|
|`np.float64`|`np.double`|

All *aliases* can be found at https://numpy.org/doc/stable/reference/arrays.scalars.html#sized-aliases

### Trigonometric functions

## `np.savetxt` and `np.loadtxt`

```python
array_2d = np.vstack([array_1, array_2])
array_2d
```

```python
np.savetxt(fname = 'array_csv.csv', X = array_2d)
```

```python
!ls
```

```python
!head array_csv.csv
```

```python
np.savetxt(fname = 'array_csv.csv', X = array_2d, fmt = '%s')
```

```python
!head array_csv.csv
```

```python
np.savetxt(fname = 'array_csv.csv', X = array_2d, fmt = '%s', delimiter = ',')
!head array_csv.csv
```

```python
array_csv_loaded = np.loadtxt(fname = '/content/array_csv.csv', dtype=int, delimiter = ',')
array_csv_loaded
```

 `np.loadtxt` had other interesting parameters, such as:
- `skiprows`: number of rows to skip from the beginning of the file
- `usecols`: column indexes to load
- `max_rows`: maximum number of rows to load

Other methods are available according to the documentation: https://numpy.org/devdocs/reference/routines.io.html#input-and-output
